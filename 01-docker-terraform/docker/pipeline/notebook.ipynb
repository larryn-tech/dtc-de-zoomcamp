```python
import pandas as pd
```

```python
# Import NYC Taxi dataset
prefix = 'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/yellow/'

dtype = {
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": "float64",
    "RatecodeID": "Int64",
    "store_and_fwd_flag": "string",
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "payment_type": "Int64",
    "fare_amount": "float64",
    "extra": "float64",
    "mta_tax": "float64",
    "tip_amount": "float64",
    "tolls_amount": "float64",
    "improvement_surcharge": "float64",
    "total_amount": "float64",
    "congestion_surcharge": "float64"
}

parse_dates = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime"
]

df = pd.read_csv(
    prefix + 'yellow_tripdata_2021-01.csv.gz',
    nrows=100,
    dtype=dtype,
    parse_dates=parse_dates
)
```

# Ingesting Data into Postgres

```python
# Add progress bar
!uv add tqdm

from tqdm.auto import tqdm
```

```python
# Install SQLAlchemy to allow data ingestion into Postgres
!uv add sqlalchemy "psycopg[binary,pool]"
```

```python
# Create db connection
from sqlalchemy import create_engine
engine = create_engine('postgresql+psycopg://root:root@localhost:5432/ny_taxi')
```

```python
# Get DDL Schema
print(pd.io.sql.get_schema(df, name='yellow_taxi_data', con=engine))
```

```python
# Create the table without adding data
df.head(n=0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
```

```python
# Ingest the data in chunks
df_iter = pd.read_csv(
    prefix + 'yellow_tripdata_2021-01.csv.gz',
    dtype=dtype,
    parse_dates=parse_dates,
    iterator=True,
    chunksize=100000
)

total_chunks = 0
for chunk in df_iter:
    total_chunks += 1
    print(len(chunk))
```

# Complete Ingestion Loop

```python
first = True

df_iter = pd.read_csv(
    prefix + 'yellow_tripdata_2021-01.csv.gz',
    dtype=dtype,
    parse_dates=parse_dates,
    iterator=True,
    chunksize=100000
)

for df_chunk in tqdm(df_iter, total=total_chunks):

    if first:
        # Create table schema (no data)
        df_chunk.head(0).to_sql(
            name="yellow_taxi_data",
            con=engine,
            if_exists="replace"
        )
        first = False
        print("Table created")

    # Insert chunk
    df_chunk.to_sql(
        name="yellow_taxi_data",
        con=engine,
        if_exists="append"
    )

    print("Inserted:", len(df_chunk), "rows")

print("Insertion completed")
```
